```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
y = pd.DataFrame()
y['price'] = df['price']
x = df.drop(['price'], axis = 1)
df1, df2, df3, df4 = train_test_split(x, y, test_size = 0.99, random_state = 0)
```

```python
df1['Price'] = df['price']
df1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (10,8))
sns.regplot(df1.sqft_living, df1.Price)
plt.title('Square Feet vs. Housing Price', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.regplot(df1.sqft_living, df1.Price, data = df1, color = 'r')
plt.title('Square Feet vs. Housing Price', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.regplot(df1.sqft_living, df1.Price, data = df1, scatter_kws = {'color': 'purple', 'alpha': 0.3}, line_kws = {'color': '#CCCC00', 'alpha': 0.3, 'lw':6})
plt.title('Square Feet vs. Housing Price', fontsize = 20)
plt.show()
```
